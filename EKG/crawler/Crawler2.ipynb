```python
import pandas as pd
import requests,json,re
from bs4 import BeautifulSoup
from selenium import webdriver
import re
import tqdm
```

```python
url = 'http://finance.eastmoney.com/a/202008311613443563.html'
```

```python
texts = []
extras = ['原文标题','浪客直播','文章主体','责任编辑']
```

```python
def extract_news(url):
    r = requests.get(url)
    r.encoding = 'utf-8'
    demo  = r.text
    soup = BeautifulSoup(demo,'html.parser')  # BeautifulSoup有好几种解析器，我们使用html.parser解析html页面
    content = soup.find(id='ContentBody').contents
    for div in content:
        p = BeautifulSoup(str(div)).get_text(separator="") 
        if p != '' and p not in extras and re.search(r'责任编辑',p) == None and re.search(r'重要公告',p) == None and re.search(r'年半年报',p) == None:
            p = p.replace('\u3000','')
            texts.append(p)
```

```python
lines = sum([1 for i in open("links-july.txt", "r")])
with open('links-july.txt','r') as f:
    for idx,link in tqdm.tqdm(enumerate(f),total=lines):
        extract_news(link)
```

```python
with open('news-july.txt','w') as f:
    f.write('\n\n'.join(texts))
```
