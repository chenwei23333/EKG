```python
import pandas as pd
import json
```

```python
with open("train_data_me.json",'r') as f:
    load_dict = json.load(f)
    print(len(load_dict))
```

```python
count = 0 
new_dict = []
for i in range(173095):
    for tri in load_dict[i]['spo_list']:
        if '总部地点' in tri or '注册资本' in tri or '成立日期' in tri or '创始人' in tri or '董事长' in tri:
            count += 1
            new_dict.append(load_dict[i])
            break
print(count)
```
